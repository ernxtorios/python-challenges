### Funciones de agregaciÃ³n
Las funciones de agregaciÃ³n son funciones que toman una serie de entradas y devuelven una Ãºnica salida.

Las mÃ¡s comunes que utilizamos en Pandas son las que toman series numÃ©ricas y devuelven un estadÃ­stico de interÃ©s, como la media.

```python
print(df_movies["rating"].mean())
print(df_movies["rating"].min())
print(df_movies["rating"].max())
```

El mÃ©todo describe() proporciona un informe de estadÃ­sticas resumidas sobre una serie numÃ©rica dada

```python
df_movies["rating"].describe()
```

TambiÃ©n podemos usar este mÃ©todo en un DataFrame para verlo aplicado a cada columna numÃ©rica.

```python
df_movies[["rating", "rotton_tomato"]].describe()
```

```python
df_movies[["rating", "rotton_tomato"]].describe().round(2)
```

### Agrupaciones
A veces, querremos ver los resultados de una funciÃ³n de agregaciÃ³n por categorÃ­a en una columna no numÃ©rica. Para ello, primero usamos el mÃ©todo groupby([columns]) y especificamos la categorÃ­a por la que queremos agrupar.

```python
df_movies.groupby(["genres"])
```

```python
df_movies_per_genres = df_movies.groupby(["genres"])[["movie"]].count()

df_movies_per_genres
```

```python
df_movies_per_genres_top_ten = df_movies.groupby(["genres"]).count().head(10)[["movie"]]

df_movies_per_genres_top_ten
```

### Reset index
Normalmente, los Ã­ndices de fila son nÃºmeros, pero tambiÃ©n pueden tener nombre.

```python
df_movies_per_genres.columns
```

Para restablecer el Ã­ndice a nÃºmeros y mover el Ã­ndice actual a una columna, usamos reset_index()

```python
df_movies_per_genres_reset = df_movies_per_genres.reset_index()

df_movies_per_genres_reset
```

### OrdenaciÃ³n
Pandas proporciona un mÃ©todo sort_values() para ordenar las filas en un DataFrame o Series.

Si se llama en un DataFrame, debe especificar por quÃ© columna estÃ¡ ordenando, como esto: df.sort_values([col]).

```python
df_movies.sort_values(["rating"])
```

Si se aplica a una Serie solo hay una columna, por lo que no es necesario especificar

```python
df_movies_per_genres["movie"].sort_values()
```

Por defecto, sort_values() ordena en orden ascendente. Puede especificar el parÃ¡metro "ascending=False" para cambiarlo a orden descendente.

```python
df_movies_per_genres["movie"].sort_values(ascending=False)
```

### NaN
QuizÃ¡s hayas notado que nuestro DataFrame contiene valores NaN. Estos indican un valor faltante.

Existen varias maneras de gestionar los valores faltantes. A menudo, la presencia de estos valores causa problemas en las tareas computacionales.

Primero, podemos comprobarlo usando el mÃ©todo isna(), alias de isnull(), y el mÃ©todo sum() para contar el nÃºmero de valores NaN presentes.

```python
df_movies_nan = df_movies[["movie", "title", "genres", "rating", "rotton_tomato"]]

df_movies_nan
```

```python
df_movies_nan.isna().sum()
```

### Dropping NaN
Una forma de gestionar NaN es descartar todas las filas que tengan valores NaN. Podemos usar el mÃ©todo dropna() para ello.

```python
df_movies_nan.dropna()
```

### Impute columns
Sin embargo, al eliminar registros, se descarta mucha informaciÃ³n.

En lugar de eliminar filas con valores faltantes, podemos imputar los valores faltantes usando fillna() y especificando un valor predeterminado.

```python
df_movies_nan.fillna("Missing")
```

A menudo, queremos imputar diferentes valores a distintas columnas. Por ejemplo, con valores numÃ©ricos, podemos imputar con la media, la mediana, etc. Para caracterÃ­sticas categÃ³ricas, es comÃºn imputar con la moda o una categorÃ­a especial.

Podemos pasar un diccionario a fillna() que tenga como clave los nombres de las columnas y como valor el valor con el que se imputarÃ¡ la columna.

Opcionalmente, puede especificar inplace=True si desea actualizar el DataFrame subyacente.

```python
df_movies_nan.fillna({"rating": 0, "rotton_tomato": 50}, inplace=False)
```
